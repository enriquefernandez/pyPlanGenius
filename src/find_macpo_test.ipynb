```python
%load_ext autoreload
%autoreload 2
from pddl import parser
from pddl.pddl_world import *
from find_macpo import find_macpo
```

```python
domain = parser.parse_domain_file('../pddl/simpleworld-domain.pddl')
problem = parser.parse_problem_file('../pddl/simpleworld-problem.pddl')
print(domain)
print(problem)
```

```python
opa = domain.operator_schemas['operator-a']
opb = domain.operator_schemas['operator-b']
opc = domain.operator_schemas['operator-c']
opd = domain.operator_schemas['operator-d']
ope = domain.operator_schemas['operator-e']
opf = domain.operator_schemas['operator-f']
x = problem.objects[0].name
```

```python
iniState = State(problem.initial_conditions.predicates)
endState = State(problem.goal_conditions)
```

```python
from IPython.core.display import SVG, Image
from IPython.display import display, display_svg, display_png
from plan_graphs import PartialOrderGraph, TotalOrderGraph
#pograph = PartialOrderGraph(plan)
#display(pograph)
```

```python
# Creating an action sequence and computing states (Total Order Plan)
asec = ActionSequence(iniState, endState)
state_sequence = [iniState]
state_sequence.append(asec.add_action_step(ActionStep(opa, x),state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(opb, x),state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(opc, x),state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(opd, x),state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(opf, x),state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(ope, x),state_sequence[-1]))
total_order_graph = TotalOrderGraph(asec)
print(asec)
display(total_order_graph)
```
