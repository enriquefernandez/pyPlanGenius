```python
%load_ext autoreload
%autoreload 2
from pddl import parser
from pddl.pddl_world import *
from find_macpo import find_macpo
from IPython.core.display import SVG, Image
from IPython.display import display, display_svg, display_png
from plan_graphs import PartialOrderGraph, TotalOrderGraph
```

```python
domain = parser.parse_domain_file('../pddl/balls-domain.pddl')
problem = parser.parse_problem_file('../pddl/balls-problem.pddl')

find_ball = domain.operator_schemas['find-ball']
pickup_ball = domain.operator_schemas['pickup-ball']
drop_ball = domain.operator_schemas['drop-ball']

yellowball = problem.get_object('yellowball').name
greenball = problem.get_object('greenball').name
redbasket = problem.get_object('redbasket').name
pr2 = problem.get_object('pr2').name


iniState = State(problem.initial_conditions.predicates)
endState = State(problem.goal_conditions)

# Creating an action sequence
asec = ActionSequence(iniState, endState)
state_sequence = [iniState]
state_sequence.append(asec.add_action_step(ActionStep(find_ball, [pr2,yellowball]), state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(pickup_ball, [pr2,yellowball]), state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(drop_ball, [pr2,yellowball,redbasket]), state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(find_ball, [pr2,greenball]), state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(pickup_ball, [pr2,greenball]), state_sequence[-1]))
state_sequence.append(asec.add_action_step(ActionStep(drop_ball, [pr2,greenball,redbasket]), state_sequence[-1]))

total_order_graph = TotalOrderGraph(asec)
#print(asec)
#display(total_order_graph)
#dot_code = total_order_graph.dot_graph.to_string()
total_order_graph.save_pdf('output_graphs/balls-to-limitations.pdf')
```

```python
# Find Partial Order Graph (no threat resolution)
po_plan_nothreats = find_macpo(asec, state_sequence, False)
print(po_plan_nothreats)
po_graph_nothreats = PartialOrderGraph(po_plan_nothreats)
display(po_graph_nothreats)
po_graph_nothreats.save_pdf('output_graphs/balls-po-limitations.pdf')
po_graph_nothreats.dot_graph.write_dot('output_graphs/balls-po-limitations.dot')
```

```python
# Find Partial Order Graph (threat resolution)
po_plan_threats = find_macpo(asec, state_sequence, True)
print(po_plan_threats)
po_graph_threats = PartialOrderGraph(po_plan_threats)
display(po_graph_threats)
po_graph_threats.save_pdf('output_graphs/balls-po-limitations-threats.pdf')
```
